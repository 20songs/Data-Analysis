### 앞, 뒤 50줄 삭제하기

```python
f = open('C:/Users/admin/Desktop/yes24/information/stopwords.txt','r',encoding='utf-8')
stopwords = []
for line in f.readlines():
    stopwords.append(line.rstrip())
f.close()
```

```python
file_list = os.listdir('C:/Users/admin/Desktop/yes24/data/337')

print ("file_list: {}".format(file_list))
```

```python
len(file_list)
```

```python
from soynlp.normalizer import *
```

```python
front50 = []
back50 = []

for txt in file_list:
    file = open('C:/Users/admin/Desktop/yes24/data/337/{}'.format(txt),'r',encoding='utf-8')
    sentence = file.readlines()
    file.close()
    
    new_sen = []
    
    for line in sentence:
        line = only_hangle_number(line)
        if len(line)==0:
            del(line)
        else:
            new_sen.append(line)
    
    token = []
    
    for lin in new_sen:
        word = lin.split()
        word = [words for words in word if not words in stopwords] # 불용어 제거
        token.append(word)
    
    front50.append(token[:50])
    back50.append(token[-50:])
```
