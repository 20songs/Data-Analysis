```python
import pandas as pd
import numpy as np
```

```python
users = pd.read_csv('./accounts_context_add_prefer_0624.csv',encoding='utf-8')
```

```python
users.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
users.drop('Unnamed: 0.1',axis=1,inplace=True)
```

```python
orders = pd.read_csv('./order_newpreference.csv',encoding='utf-8')
```

```python
orders.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
orders
```

```python
users
```

```python
grouped = orders.groupby('account_id')
```

```python
new_pre = grouped.sum()
```

```python
new_pre['new_pref_2'] = new_pre['new_pref'].apply(lambda x:0 if x==0 else 1)
```

```python
new_pre.reset_index(inplace=True)
```

```python
new_pre
```

```python
add_pre = new_pre[['account_id','new_pref_2']]
```

```python
add_pre
```

```python
temp_context = users.merge(add_pre,how='left')
```

```python
temp_context.head()
```

```python
rfm = pd.read_csv('./rfm_0624.csv',encoding='utf-8')
```

```python
add_rfm = rfm[['account_id','monetary','frequency','recency']]
```

```python
add_rfm.isnull().sum()
```

```python
users_context = temp_context.merge(add_rfm,how='left')
```
